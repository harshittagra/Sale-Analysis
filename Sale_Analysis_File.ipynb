# Exploratory Data Analysis

# ON BASIS OF GENDER

```python
df
```

Now we will draw a count plot which will show us that how many married , unmmaried man and woman has ordered a product in this sale

```python
x= sns.countplot(data=df,x='Gender',hue='Marital_Status')
for bars in x.containers:
  x.bar_label(bars)

```

```python
 a = df.groupby(['Gender'],as_index=False)[['Amount']]
```

```python
sales_gen = a.sum()
```

```python
sales_gen
```

```python
sns.catplot(data = sales_gen,x='Gender',y='Amount',kind='bar')
```

**From above two graphs we can see that most of the buyers are Female and purchasing power of Female is more than male .**

# Occupation
